## Cross Product
Given two vectors $\vec{a}$ and $\vec{b}$, the cross product $a \times b$ is:
\begin{align*}
a \times b=\left|
\begin{matrix}
i & j & k \\
a_1 & a_2 & a_3 \\
b_1 & b_2 & b_3 \\
\end{matrix}\right|=
\left|
\begin{matrix}
a_2 & a_3  \\
b_2 & b_3 \\
\end{matrix}\right|i-
\left|
\begin{matrix}
a_1 & a_3  \\
b_1 & b_3 \\
\end{matrix}\right|j+
\left|
\begin{matrix}
a_1 & a_2  \\
b_1 & b_2 \\
\end{matrix}\right|k
\end{align*}

Then calculate the determinants as described before:
\begin{align*}
=(a_2b_3-a_3b_2)i-(a_1b_3-a_3b_1)j+(a_1b_2-a_2b_1)k
\end{align*}

Example:<br>
\begin{align*}
\vec{a}=\begin{bmatrix} 1 \\ 3 \\ 4 \end{bmatrix} \ , \ \vec{b}=\begin{bmatrix} 2 \\ 7  \\ -5 \end{bmatrix}
\end{align*}
\begin{align*}
a \times b=\left|
\begin{matrix}
i & j & k \\
1 & 3 & 4 \\
2 & 7 & -5 \\
\end{matrix}\right|=
\left|
\begin{matrix}
3 & 4  \\
7 & -5 \\
\end{matrix}\right|i-
\left|
\begin{matrix}
1 & 4  \\
2 & -5 \\
\end{matrix}\right|j+
\left|
\begin{matrix}
1 & 3  \\
2 & 7 \\
\end{matrix}\right|k
\end{align*}

\begin{align*}
&=((3)(-5)-(4)(7))i-((1)(-5)-(4)(2))j+((1)(7)-(3)(2))k \\
\\
&=(-15-28)i-(-5-8)j+(7-6)k\\
\\
&= -43i+13j+k
\end{align*}
The cross product will always be orthogonal to the plane containing the two original vectors.

```python
# With Python
# Create two vectors
a = np.array([1, 3, 4])
b = np.array([2, 7, -5])

# Calculate the cross product
cross_product = np.cross(a, b)
cross_product
```

```python
# Create a figure and 3D axes
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

# Plot the original vectors
ax.quiver(0, 0, 0, a[0], a[1], a[2], color='b', label='a', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, b[0], b[1], b[2], color='g', label='b', arrow_length_ratio=0.1)

# Plot the cross product vector
ax.quiver(0, 0, 0, cross_product[0], cross_product[1], cross_product[2], color='r', label='a x b',arrow_length_ratio=0.1)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_xlim([-30,10])
ax.set_ylim([-10,20])
ax.set_zlim([-10,10])
ax.set_title('Cross Product')
ax.legend()

plt.show()
```

Properties of cross products:<br>
1. The cross product of a vector with itself will always be $0$ denoted:
   \begin{align*}
    a \times a =0
   \end{align*}
2. The magnitude of a cross product is:
    \begin{align*}
    |a \times b|=|a||b|\sin \theta
   \end{align*}
   where $|a|$ is the magnitude of $a$ and $|b|$ is the magnitude of $b$.<br>
   <br>
3. Any two parrallel vectors have a cross product of $0$ because the angle between them is $0$.<br>
<br>
4. Not commutative:
\begin{align*}
a \times b \not = b \times a \ , \ a \times b = -b \times a
\end{align*}
5. Not associative:
\begin{align*}
(a \times b) \times c\not = a \times (b \times c) 
\end{align*}
6. Distributive:
\begin{align*}
a \times (b + c)=(a \times b)+(a\times c)
\end{align*}
